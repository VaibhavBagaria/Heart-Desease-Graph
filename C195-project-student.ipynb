# Task 1 - Find the diabetic and hight blood pressure patients across all age groups, and conclude the risk heart stroke is higher in which age group

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df = pd.read_csv('heart_disease.csv')
df
```

```python
diabetes=df.loc[df['diabetes'] == 1]
diabetes
```

```python
groupby_diabetes=diabetes.groupby('age')['diabetes'].count().reset_index()
groupby_diabetes
```

```python
HBP=df.loc[df['high_blood_pressure'] == 1]
HBP
#Filter by high_blood_pressure(condition will be who has high_blood_pressure) and create new dataframe

```

```python
groupby_HBP=HBP.groupby('age')['high_blood_pressure'].count().reset_index()
groupby_HBP
```

```python
plt.figure(figsize=(12,8))

plt.scatter(groupby_diabetes['age'],groupby_diabetes['diabetes'], label='People with Diabetes',color='aqua',alpha=0.5,
            s=groupby_diabetes['diabetes']*100)
plt.scatter(groupby_HBP['age'],groupby_HBP['high_blood_pressure'], label='People with High Blood Pressure',color='blue',alpha=0.5,
            s=groupby_HBP['high_blood_pressure']*100)

plt.xlabel('Age')
plt.ylabel('People with 10 years of CHD')
plt.title('Different ages with 10 years of CHD')
plt.legend()
plt.show()
```

# Task 2 - Find as per gender who has not normal platelets level in blood

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df2 = pd.read_csv('heart_disease.csv')
platelet=df2.loc[(df2['platelets'] < 150000.00) | (df2['platelets'] > 450000.00)]
platelet
```

```python
groupby_platelet=platelet.groupby('gender')['platelets'].count().reset_index()
groupby_platelet
```

```python
#Plot a pie chart as per the gender to show the percentage of male and female who has not normal platelets
value = groupby_platelet['platelets']
label = groupby_platelet['gender']
plt.pie(value,labels=label, autopct='%0.2f%%', radius=2)
plt.title('Different genders with unnormal platelets')
plt.show()
```

Conclusion - Gender 1
